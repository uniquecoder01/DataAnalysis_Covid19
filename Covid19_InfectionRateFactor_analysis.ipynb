```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

### Importing dataset

```python
corona_dataset_df = pd.read_csv(r"C:\Users\prabh\Desktop\Covid19-Data_Analysis\csv\covid19_Confirmed_dataset.csv")
corona_dataset_df
```

### Shape of datadrame

```python
corona_dataset_df.shape
```

### Deleting useless columns

```python
corona_dataset_df.drop(["Lat","Long"], axis = 1, inplace = True)
corona_dataset_df.head(10)
```

### Aggregating rows by the country

```python
new_corona_df = corona_dataset_df.groupby("Country/Region").sum()
new_corona_df.head(10)
```

### Shape of aggregated dataset

```python
new_corona_df.shape
```

### Calculating good measure

```python
new_corona_df.loc["India"].plot()
```

### Plot of cases in India for first 3 days

```python
new_corona_df.loc["India"][:3].plot()
```

### Calculating and plotting first derivative of the curve

```python
new_corona_df.loc["India"].diff().plot()
```

### Finding maximum infection rate for India

```python
new_corona_df.loc["India"].diff().max()
```

### Finding maximum infection rate for all Countries

```python
countries = list(new_corona_df.index)
max_infection_rate = []

for c in countries:
    max_infection_rate.append(new_corona_df.loc[c].diff().max())

new_corona_df["max_infection_rate"] = max_infection_rate
    
```

```python
new_corona_df.head()
```

### Creating new Dataframe with only needed columns

```python
corona_data = pd.DataFrame(new_corona_df["max_infection_rate"])
```

```python
corona_data.head()
```

```python
corona_data.shape
```
